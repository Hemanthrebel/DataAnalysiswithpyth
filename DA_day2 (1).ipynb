```python
#plot the scores of 5 different subjects?
#subjects as x-axis
#marks as y-axis
stu1=[88,74,99,96,91]
stu2=[78,79,89,92,95]
sub=["maths","physics","chemistry","biology","english"]
plt.plot(sub,stu1,label="stu1",color="y")
plt.plot(sub,stu2,label="stu2",color="k")
plt.title("marks of subjects")
plt.legend()
```

```python
#create subplots(bar)showing the profits of two companies
y=np.array(["2019-2020","2020-2021","2021-2022","2022-2023"])
a=np.array([230,560,780,127,128])
a1=np.diff(a)
print(a1)
b=np.array([200,160,270,127,400])
b1=np.diff(b)
print(b1)
plt.subplot(1,2,1)
plt.bar(y,a1,label="profit a",color="k")
plt.legend()
plt.subplot(1,2,2)
plt.bar(y,b1,label="profit b",color="c")
plt.legend()


```
