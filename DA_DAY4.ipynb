```python
url=("https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1")
req=requests.get(url)
a=BeautifulSoup(req.text,'html')
print(a)
#<th>=table header
#<tr>=table row
#<td>=table data
```

```python
table=a.find("table")
print(table)
```

```python
heading=table.findAll("th")
print(heading)
```

```python
import pandas as pd
```

```python
rows=table.findAll("tr")
print(rows)
```

```python
data=table.findAll("td")
print(data)
```

```python
head=[i.text for i in heading]
print(head)
```

```python
for i in rows[1:9]:
  rd=i.findAll("td")
  eachrow=[i.text for i in rd]
  print(eachrow)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=head)
print(df)
a=0
for i in rows[1:9]:
  er=i.find_all("td")
  eachrow=[j.text for j in er]
  df.loc[a]=eachrow
  a=a+1
df
```

```python
df.to_csv("naruto.csv")
```
